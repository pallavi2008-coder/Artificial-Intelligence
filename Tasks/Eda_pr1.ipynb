```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
ds=pd.read_csv("employee_dataset_for_eda.csv")
ds
```

```python
ds.head()
```

```python
ds.info()
```

```python
ds.describe()
```

```python
sns.countplot(data=ds,x="Department", palette='Set1')
plt.title("Employee count per Department")
```

```python
sns.countplot(data=ds,x="Gender",palette='Set2')
plt.title("Gender Distribution")
```

```python
sns.countplot(data=ds,x='Location',palette='dark')
plt.title("Number of location")
plt.xticks(rotation=45)
```

```python
sns.histplot(ds['Age'],bins=10,kde=True)
plt.title("Age Distribution")
```

```python
sns.histplot(ds['Years_of_Experience'],bins=15,kde=True)
plt.title("Years of Experience")
```

```python
sns.barplot(data=ds,y="Monthly_Salary",color='lightblue')
plt.title("Salary spread")
```

```python
sns.barplot(data=ds,x="Department",y="Monthly_Salary",estimator=np.mean, palette='magma')
plt.title("Average Salary by Department")
```

```python
custom_colors = ['#FF6F61', '#6B5B95', '#88B04B']
sns.boxplot(data=ds, x="Department", y="Performance_Score",palette=custom_colors)
plt.title("Performance score by department")
```

```python
sns.violinplot(data=ds,x="Gender",y="Monthly_Salary",palette='muted')
plt.title("Salary by Gender")
```

```python
sns.heatmap(ds[['Age','Years_of_Experience','Monthly_Salary','Performance_Score']].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
```

```python
sns.scatterplot(data=ds,x="Years_of_Experience",y="Monthly_Salary",hue="Department")
plt.title("Salary vs Experience scatterplot")
```
