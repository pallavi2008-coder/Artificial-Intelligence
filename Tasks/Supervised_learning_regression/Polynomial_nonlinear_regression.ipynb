```python
#Importing libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Importing and reading dataset

ds=pd.read_csv("Position_Salaries.csv")
ds
```

```python
x=ds[['Level']]
y=ds[['Salary']]
```

```python
y
```

```python
#Plot graphs

plt.scatter(x,y)
plt.title("Scatter Plot")
plt.show()
```

```python
#Model Fitting

from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
poly_reg=PolynomialFeatures(degree=5)
x_poly=poly_reg.fit_transform(x)
reg = LinearRegression()
reg.fit(x_poly,y)
```

```python
#Visualization

plt.scatter(x,y)
plt.plot(x,reg.predict(poly_reg.transform(x)),color='red')
plt.show()
```

```python
#Predict for my input

myinput=pd.DataFrame(data=[[11]],columns=["Level"])
result=reg.predict(poly_reg.transform(myinput))
result
```

```python
#Predict for user input

level=float(input("Enter level: "))
myinput=pd.DataFrame(data=[[level]], columns=["Level"])
poly_input= poly_reg.transform(myinput)
result=reg.predict(poly_input)
print(f"Predicted result for level {level}: {round(result[0,0])}")
```
