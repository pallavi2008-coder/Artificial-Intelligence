```python
import seaborn as sns
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
ds=pd.read_csv("product_data.csv")
ds
```

```python
print(ds.to_string())
```

```python
ds.info()
```

```python
ds.loc[2,'Price']=45000
ds.loc[5,'Quantity']=2
```

```python
ds['Price']=pd.to_numeric(ds['Price'])
ds['Quantity']=pd.to_numeric(ds['Quantity'])
```

```python
ds['PurchaseDate']=pd.to_datetime(ds['PurchaseDate'])
```

```python
ds.info()
```

```python
x=ds['Price'].mean()
ds['Price'].fillna(x,inplace=True)
```

```python
ds.info()
```

```python
x=ds['Quantity'].median()
ds['Quantity'].fillna(x,inplace=True)
```

```python
ds.info()
```

```python
print(ds.to_string())
```

```python
ds['Price']=ds['Price'].round(2)
ds['Quantity']=ds['Quantity'].round(0)
```

```python
print(ds.to_string())
```

```python
ds['Quantity']=ds['Quantity'].astype(int)
```

```python
print(ds.to_string())
```

```python
ds.to_csv("Cleaned_data_pr2.csv",index=False)
```
