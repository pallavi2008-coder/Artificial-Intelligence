```python
#Import Libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
#Import dataset

ds=pd.read_csv("Salary_Data.csv")
ds
```

```python
#Divide dataset as Independent(input) and Dependent(output) variables

x=ds[['YearsExperience']]
y=ds[['Salary']]
```

```python
ds.columns
```

```python
plt.scatter(x,y)
plt.title("Scatter Plot")
plt.show()
```

```python
#Splitting dataset as training and testing

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.2, random_state=0)
```

```python
x_test
```

```python
#Define Model

from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(x_train,y_train)
```

```python
#Predict

result=reg.predict([[9]])
result
```

```python
result=reg.predict(x_test)
result
```

```python
y_test
```

```python
#Visualize 

plt.scatter(x,y)
result=reg.predict(x)
plt.plot(x,result,color='red')
plt.show()
```

```python
#For user input

years = int(input("Enter years of experience: "))
my_input=pd.DataFrame(columns=['YearsExperience'],data=[[years]])
result=reg.predict(my_input)
print("Your Salary must be close to",round(result[0,0]))
```

```python
from sklearn.metrics import r2_score
y_pred=reg.predict(x_test)
score= r2_score(y_test,y_pred)
score
```
